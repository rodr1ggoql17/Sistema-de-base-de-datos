# Para que este codigo funcione debemos incluir los archivos 'Players.csv' y 'Teams.csv'

```python
import pandas as pd
```

```python
# Leer los jugadores y equipos en dataframes [Players.csv] y [Teams.csv].
df_players = pd.read_csv('Players.csv')
df_teams = pd.read_csv('Teams.csv')
display(df_teams)
```

```python
# Qué jugador en un equipo con "ou" en el nombre, jugó mas de 190 minutos e hizo más de 100 pases? Imprima el apellido del jugador.
df_surnames = df_players[df_players.surname.str.contains('ou')&(df_players.minutes > 190)&(df_players.passes > 100)]
df_surnames
```

```python
# ¿Cuál es el promedio, máximo y mínimo de pases de los defensas y delanteros?
print('Delanteros')
df_forward = df_players[df_players.position == 'forward']
print('Promedio pases: ', df_forward['passes'].mean())
print('Max pases: ', df_forward['passes'].max())
print('Min pases: ', df_forward['passes'].min(),'\n')

print('Defensas')
df_defender = df_players[df_players.position == 'defender']
print('Promedio pases: ', df_defender['passes'].mean())
print('Max pases: ', df_defender['passes'].max())
print('Min pases: ', df_defender['passes'].min())
```

```python
'''
¿Qué equipo tiene el mayor ratio de 'goalsFor' y 'goalsAgainst'? Imprimir sólo el nombre del equipo.
¿Cuántos jugadores que juegan en un equipo con ranking <5 jugaron más de 400 minutos?
Calcule por cada equipo la diferencia entre el jugador que da menos pases con el que da más pases. (use función groupby, agg con función lambda)
'''
goalsFor_max = df_teams['goalsFor'].max()
df_aux = df_teams[df_teams.goalsFor==goalsFor_max]
print('El equipo con más goalsFor es: ', df_aux.team)

goalsAgainst_max = df_teams['goalsAgainst'].max()
df_aux2 = df_teams[df_teams.goalsAgainst==goalsAgainst_max]
print('El equipo con más goalsAgainst es: ', df_aux2.team)
```

```python
# ¿Cuántos jugadores que juegan en un equipo con ranking < 5 jugaron más de 400 minutos?
df_united = pd.merge(df_players,df_teams,how='left')
df_united.head(20)
```

```python
df_united[(df_united.ranking < 5)&(df_united.minutes > 400)]
```
