```python
import pandas as pd
import numpy as np
```

# Archivo Excel

```python
# archivo excel
df_cars = pd.read_excel('Cars.xlsx')
df_cars
```

```python
# incluir nueva fila en el registro
new = {'id': 5,'cars':'Mazda','max_speed':130,'color':'red'}
df_cars = df_cars.append(new,ignore_index = True)
df_cars
```

```python
#promedio de velocidad maxima
print('Promedio velocidad maxima del dataframe es: ', df_cars.max_speed.mean())
```

```python
# filtrar los autos color rojo
df_cars[df_cars.color == 'red']
```

# Archivo Csv


```python
# lectura archivo csv y exportar a json
df_cars_csv = pd.read_csv('cars_csv.csv',delimiter=';')
df_cars_csv.to_json('file.json', orient = 'split', compression = 'infer', index = 'true')
```

```python
#promedio de velocidad maxima
print('Promedio velocidad maxima del dataframe es: ', df_cars_csv.max_speed.mean())
```

```python
# filtrar los autos color rojo
df_cars_csv[df_cars.color == 'red']
```

# Archivo Json

```python
#lectura
df_iris = pd.read_json('iris.json')
df_iris
```

```python
# guardar un df solo con especies 'setosa' y contar los registros
df_setosa = df_iris[(df_iris.species == 'setosa')]
df_setosa
```

```python
# total registros
print('registros de setosa:', len(df_setosa.index))
```
